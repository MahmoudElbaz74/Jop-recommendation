```python
import numpy as np
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('newData2.csv').drop('Unnamed: 0',axis=1)
df
```

```python
df['mex'] = df['job_title'].map(str) + ' ' + df['description']
df
```

```python
from nltk.corpus import stopwords
import re
import string
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
stop = stopwords.words('english')
stop_words_ = set(stopwords.words('english'))
wn = WordNetLemmatizer()

def black_txt(token):
    return  token not in stop_words_ and token not in list(string.punctuation)  and len(token)>2   
  
def clean_txt(text):
  clean_text = []
  clean_text2 = []
  clean_text = [ wn.lemmatize(word, pos="v") for word in word_tokenize(text.lower()) if black_txt(word)]
  clean_text2 = [word for word in clean_text if black_txt(word)]
  return " ".join(clean_text2)
```

```python
df['mex'] = df['mex'].apply(clean_txt)
```

```python
vec = CountVectorizer(binary=False)
features = vec.fit_transform(df['mex'])
```

```python
df2 = pd.read_csv('aplcdata2.csv').drop('Unnamed: 0',axis=1)
df2
```

```python
df2['mex2']=df2['Position'].map(str) + ' ' + df2['skills'] + ' ' + df2['city'] + ' ' + df2['Empl_type']
df2.drop(['Position','skills','city','Empl_type',],axis=1,inplace=True)
df2
```

```python
def get_app_id(h):
    h = 2
    index = np.where(df2['Applicant.ID'] == h)[0][0]
    user_q = df2.iloc[[index]]
    user_vec = vec.transform(user_q['mex2'])
    cos_similarity_vec = map(lambda x: cosine_similarity(user_vec, x),features)
    output2 = list(cos_similarity_vec)
    return output2,h
output2,h =get_app_id(int(2))
```

```python
def get_recommendation(top, df_all, scores):
  recommendation = pd.DataFrame(columns = ['ApplicantID', 'JobID',  'title', 'score'])
  count = 0
  for i in top:
      recommendation.at[count, 'ApplicantID'] = h
      recommendation.at[count, 'JobID'] = df['job_id'][i]
      recommendation.at[count, 'title'] = df['job_title'][i]
      recommendation.at[count, 'score'] =  scores[count]
      count += 1
  return recommendation
```
